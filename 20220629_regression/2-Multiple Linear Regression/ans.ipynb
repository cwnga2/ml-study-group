```python
# Data Preprocessing Template

# Importing the libraries
import numpy as np   #numpy math tool
import matplotlib.pyplot as plt #for drawing
import pandas as pd #process file

# Importing the dataset
dataset = pd.read_csv('50_Startups.csv')

dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# category : text -> 0,1,2

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [3])],   # The column numbers to be transformed (here is [0] but can be [0, 1, 3])
    remainder='passthrough'                                         # Leave the rest of the columns untouched
)
X = ct.fit_transform(X)
```

```python
# Avoiding the Dummy Variable Trap
X = X[:, 1:]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python


# b0 * 1+ b1 x1
#append 1 in column 0
X_train = np.append(arr = np.ones([40, 1]), values=X_train, axis=1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
#^ IF YOU THINK THATS ALL , DONE!
```
